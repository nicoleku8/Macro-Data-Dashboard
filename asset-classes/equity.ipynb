# Equity Index Datas

```python
import yfinance as yf
import pandas as pd

def get_categorized_futures(start="2020-01-01"):
    # Define Ticker Mapping
    standard_tickers = {'ES=F': 'SP500_Emini', 'NQ=F': 'Nasdaq_Emini', 'YM=F': 'Dow_Emini', 'RTY=F': 'Russell_Emini'}
    micro_tickers = {'MES=F': 'SP500_Micro', 'MNQ=F': 'Nasdaq_Micro', 'MYM=F': 'Dow_Micro', 'M2K=F': 'Russell_Micro'}
    
    # Combined list for one download
    all_tickers = list(standard_tickers.keys()) + list(micro_tickers.keys())
    
    # Download Close prices
    raw_data = yf.download(all_tickers, start=start)['Close']
    
    # Categorize into separate DataFrames
    standard_df = raw_data[list(standard_tickers.keys())].rename(columns=standard_tickers)
    micro_df = raw_data[list(micro_tickers.keys())].rename(columns=micro_tickers)
    
    return standard_df, micro_df

# Execute
standard_data, micro_data = get_categorized_futures()

print("--- Standard E-mini Data (Tail) ---")
print(standard_data.tail())

print("\n--- Micro E-mini Data (Tail) ---")
print(micro_data.tail())
```
