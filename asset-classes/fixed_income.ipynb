```python
import yfinance as yf
import pandas as pd

def get_fixed_income_data(start="2015-01-01"):
    # Define ETF Proxies (Cash Market)
    etf_proxies = {
        'TLT': '20Y_Treasury_ETF',
        'LQD': 'IG_Corporate_ETF',
        'HYG': 'High_Yield_Corp_ETF'
    }
    
    # Define Treasury Futures (Expectation Market)
    futures_tickers = {
        'ZF=F': '5Y_Note_Futures',
        'ZN=F': '10Y_Note_Futures',
        'ZB=F': '30Y_Bond_Futures'
    }
    
    # Download data
    all_tickers = list(etf_proxies.keys()) + list(futures_tickers.keys())
    raw_data = yf.download(all_tickers, start=start)['Close']
    
    # Rename for the Dashboard
    fi_df = raw_data.rename(columns={**etf_proxies, **futures_tickers})
    
    return fi_df

# Execute
fi_data = get_fixed_income_data()

print("--- Fixed Income Proxies & Futures (Tail) ---")
print(fi_data.tail())
```

```python

def get_fixed_income_dashboard(start="2020-01-01"):
    # Define Tickers
    cash_tickers = {'TLT': '20Y_Treasury', 'IEF': '7-10Y_Treasury', 'SHY': '1-3Y_Treasury', 'LQD': 'IG_Corp', 'HYG': 'High_Yield'}
    futures_tickers = {'ZT=F': '2Y_Fut', 'ZF=F': '5Y_Fut', 'ZN=F': '10Y_Fut', 'ZB=F': '30Y_Fut', '10Y=F': '10Y_Yield_Fut'}
    
    # Download
    all_tickers = list(cash_tickers.keys()) + list(futures_tickers.keys())
    data = yf.download(all_tickers, start=start)['Close']
    
    # Organize
    cash_df = data[list(cash_tickers.keys())].rename(columns=cash_tickers)
    futures_df = data[list(futures_tickers.keys())].rename(columns=futures_tickers)
    
    return cash_df, futures_df

cash_market, futures_market = get_fixed_income_dashboard()
```
