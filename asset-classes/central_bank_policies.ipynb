# Central Bank Policies


```python
import pandas as pd
from fredapi import Fred

# Initialize FRED (Replace with your actual API Key)
fred = Fred(api_key='YOUR_FRED_API_KEY')

# --- CATEGORY A: DAILY DATA ---
# This tracks the immediate cost of money and market yields
daily_ids = {
    'EFFR': 'Fed_Funds_Daily',
    'DGS10': '10Y_Treasury_Yield'
}

daily_df = pd.DataFrame()
for code, name in daily_ids.items():
    daily_df[name] = fred.get_series(code, observation_start='2015-01-01')

# --- CATEGORY B: MONTHLY DATA ---
# This tracks the "Big Picture" economic health (The Fed's Mandate)
monthly_ids = {
    'CPIAUCSL': 'CPI_Index',
    'UNRATE': 'Unemployment_Rate'
}

monthly_df = pd.DataFrame()
for code, name in monthly_ids.items():
    monthly_df[name] = fred.get_series(code, observation_start='2015-01-01')

# --- SUMMARY VIEW ---
print("--- Daily Market Rates (Tail) ---")
print(daily_df.tail())

print("\n--- Monthly Economic Prints (Tail) ---")
print(monthly_df.tail())
```
